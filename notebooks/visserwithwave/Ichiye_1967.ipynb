```python
import numpy
import matplotlib.pyplot as plt
```

### Show profile of eddy viscosity given Ichiye (1967) formula

```python
# The following 3 values are taken from Ichiye (1967)
#significant waveheight in meters
Hs = 0.38
# Wave period in seconds
T = 7.1
# 2*wavenumber in meters^-1
alpha = 0.1

depth = [numpy.float(x) for x in range(400)]
#z = numpy.float([1,2,3])
# Ichiye (1967) equation for vertical eddy viscosity
eta = 0.028*(Hs)**2/T
K = []
for z in depth:
    K.append(eta*numpy.exp(-alpha*z))
```

```python
plt.plot(K,depth)
plt.ylabel('depth')
plt.xlabel('vertical eddy viscosity')
plt.gca().invert_yaxis()
plt.show()
```

### Show profile of vertical derivative of eddy viscosity

```python
Kz = []
for z in depth:
    Kz.append(-alpha * eta * numpy.exp(-alpha*z))
```

```python
plt.plot(Kz,depth)
plt.ylabel('depth')
plt.xlabel('first derivative of vertical eddy viscosity')
plt.gca().invert_yaxis()
plt.show()
```
