```python
import pandas as pd
import yaml

vessel_types      = ["tanker", "atb", "barge", "cargo", "cruise",
                     "ferry", "fishing", "smallpass", "other"]
fuel_types = ['bunker','diesel']

# create dictionary of null values 

fuel_by_vessel = {}

for vessels in vessel_types:
    
    fuel_by_vessel[f'{vessels}'] = {}
    
    for fuels in fuel_types:
        
        fuel_by_vessel[f'{vessels}'][f'{fuels}'] = 0
        
# Many entries will stay zero.  Populate non-zero entries from values calcualated 
# [in this notebook](https://github.com/MIDOSS/analysis-rachael/blob/master/notebooks/graphics/Oil_transport/Oil_AsFuel_WA_outbound.ipynb)

# TANKER
fuel_by_vessel['tanker']['bunker'] = 0.64
fuel_by_vessel['tanker']['diesel'] = 0.36

# ATB
fuel_by_vessel['atb']['bunker'] = 0.38
fuel_by_vessel['atb']['diesel'] = 0.61

# TUG (same as atb)
fuel_by_vessel['barge']['bunker'] = 0.38
fuel_by_vessel['barge']['diesel'] = 0.61

# CARGO
fuel_by_vessel['cargo']['bunker'] = .97
fuel_by_vessel['cargo']['diesel'] = .03

# CRUISE (no data)
fuel_by_vessel['cruise']['bunker'] = 0.8
fuel_by_vessel['cruise']['diesel'] = 0.2

# FERRY
fuel_by_vessel['ferry']['bunker'] = 0.0
fuel_by_vessel['ferry']['diesel'] = 1.0

# FISHING
fuel_by_vessel['fishing']['bunker'] = 0.14
fuel_by_vessel['fishing']['diesel'] = 0.86

# SMALL PASS
fuel_by_vessel['smallpass']['bunker'] = 0.0
fuel_by_vessel['smallpass']['diesel'] = 1.0

# OTHER
fuel_by_vessel['other']['bunker'] = 0.32
fuel_by_vessel['other']['diesel'] = 0.68

with open(r'./fuel_by_vessel.yaml', 'w') as file:
    documents = yaml.safe_dump(fuel_by_vessel, file)
```
